```python
df= pd.read_csv('C:/Users/Asus/Downloads/VideoGamesSales.csv')

#check duplicates

#df = df[df.duplicated()]

#check no. of duplicate rows
df.shape[0]

df=df.drop_duplicates()

df.shape[0]
```

```python
#there are missing rows in region column and i want to replace all null values with 'North'

df['Region'] = df['Region'].fillna('North')

#now no. of rows should be same length as rows with no missing values 
df ['Region'].info
```

```python
#in NA_Sales col have null vlaues and then some values have ($) i need to remove it

df['NA_Sales'] = df['NA_Sales'].str.replace('$', '')

df
```

```python
#get avg value of NA_Sales and i will replace null values with average value of this col, but first need to convert col to numeric value

#to see if have null values
df[df['NA_Sales'].isnull()]

#to see if have 0 values
df[df['NA_Sales']=='0']

#okay to convert to datatype numeric 

#i can change it this way : df['NA_Sales'] = df['NA_Sales'].astype(float)

df['NA_Sales']= pd.to_numeric(df['NA_Sales'], errors = 'coerce') #errors='coerce': If a value can't be converted, it will be set to NaN.

#check datatype
#df.info()

#if have null and want to replace with mean , you can do this method or the one below
#df['NA_Sales']= df['NA_Sales'].fillna(df['NA_Sales'].mean())

df
```

```python

# i took mean and filled all NaN values with the mean 

#i redid this cell after i renamed the column NA_Sales to National Sales

df

average_sales = df['National Sales'].mean().round(2)

average_sales


df['National Sales']= df['National Sales'].fillna(average_sales)


df['National Sales']

df

```

```python
# some values are united states, United States, or USA so we need to fix this : 

df['Country']= df['Country'].str.replace('USA', 'United States')

#OR can write as : 

#df['Country']= df['Country'].replace({'USA', 'United States'})

#check
df[df['Country'] == 'USA']

#makes all first letters of each word in upper case 
df['Country']= df['Country'].str.title()

#renaming columns

df=df.rename(columns = {'NA_Sales' : 'National Sales', 'Global_Sales': 'Global Sales', 'NA_Profit':'National Profit', 'Global_Profit':'Global Profit'})

#check
df.head(2)
```

```python
#calculate the 95th percentile

sales_cap = df['National Sales'].quantile(0.95)
sales_cap

#95% of values in this column are less than this value 


#I WANT TO REMOVE OUTLIERS & replace them with sales_cap: 
#here in this code, im saying if condition is met then replace with sales_cap , otherwise keep as is
df['National Sales'] = np.where(df['National Sales'] > sales_cap , sales_cap, df['National Sales'])

df
```

```python
#create a bar chart 

National_Sales = df.groupby(['Region','Country'])['National Sales'].sum().reset_index().sort_values(by = 'National Sales', ascending = False)

National_Sales

sns.barplot(data = National_Sales , x='Region' , y = 'National Sales', hue = 'Country') #to include a third variable do it in form of hue

plt.title ('National Sales by Region & Country', fontweight = 'bold')

plt.xlabel ('Region', fontweight = 'bold')

plt.ylabel('National Sales', fontweight = 'bold')
```

```python
#do boxplot and display both mean and median in it 

sns.boxplot(data = df ,  x='Country' , y = 'National Sales', hue = 'Genre', showmeans= True, meanprops = {'marker': 'o', 'markerfacecolor': 'white', 'markersize': '3', 'markeredgecolor': 'green'})

plt.title ('National Sales by Country & Genre', fontweight = 'bold')

plt.xlabel ('Country', fontweight = 'bold')

plt.ylabel('National Sales', fontweight = 'bold')

plt.show()
```

```python
# create a pie chart 

sales = df.groupby('Country')[['Global Sales', 'National Sales']].sum().reset_index()
sales

Country = sales['Country']
National_Sales= sales['National Sales']

Global_Sales= sales['Global Sales']

#fig, axs: This unpacks the return values of plt.subplots. fig is a Figure object, and axs is an array of Axes objects. Since you're creating multiple subplots, axs will be an array (in this case, a 1D array with two elements).

#plt.subplots(1, 2, ...): The plt.subplots function creates a figure and a grid of subplots. The arguments 1, 2 specify the number of rows and columns of subplots, respectively. Here, 1 row and 2 columns mean there will be two subplots arranged horizontally.


fig, axs = plt.subplots(1, 2, figsize = (10,7))


#axs[0]: Refers to the first subplot in the array of Axes objects.
#The format '%1.1f%%' means that the percentage will be displayed with one decimal place.

axs[0].pie(National_Sales, labels =Country, autopct='%1.1f%%', startangle = 90)
axs[0].set_title('National Sales by Country', fontweight= 'bold')


axs[1].pie(Global_Sales, labels =Country, autopct='%1.1f%%', startangle = 90)
axs[1].set_title('Global Sales by Country', fontweight= 'bold')
```

```python
year_sale = df.groupby('Year')[['Global Sales', 'National Sales']].sum().reset_index()
year_sale

plt.plot(year_sale['Year'], year_sale['National Sales'], marker = 'o', linestyle='-', color= 'blue', label= 'National Sales')


plt.plot(year_sale['Year'], year_sale['Global Sales'], marker = 's', linestyle='--', color= 'brown', label= 'Global Sales')


plt.title('National & Global Sales Each Year')

plt.xlabel('Year')

plt.ylabel('Sales')

plt.legend()

plt.grid(True)

plt.show()
```
